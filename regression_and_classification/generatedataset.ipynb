Program to generate a one dimensional dataset with normal distribution as the probability density function with a  given mean and standard deviation----In a normal distribution, 68% of the data set will lie within ±1 standard deviation of the mean. 95% of the data set will lie within ±2 standard deviations of the mean. And 99.7% of the data set will lie within ±3 standard deviations of the mean-----

```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
import scipy.stats as st
```

when number of samples is increased the probability density is much smoother

```python
mean = 8.0 
std = 2.0

x = np.linspace(0, 16, 100)

y = st.norm.pdf(x,mean,std)
 
print(y)
plt.plot(x,y,color='red')


plt.title('dataset generated using normal distribution ',fontsize=20,color='r')

plt.xlabel('range')
plt.ylabel('Normal Distribution')

plt.savefig("normal_dist.png")
```

```python
plt.scatter(x,y,marker='x',label='samples')
plt.xlabel('x')
plt.ylabel('distribution')
plt.title('Scatter plot of the normally distributed samples',fontsize='25',color='blue')
plt.legend()
plt.savefig("normal_dist_scatter.png")
```
